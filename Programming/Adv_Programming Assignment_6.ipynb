```python
'''Q.1 You are given two strings s and t. String t is generated by randomly shuffling string s and then adding one more 
letter at a random position Return the letter that was added to t.'''

def find_the_difference(str1, str2):
    ''' '''
    result = ''
    for i in str2:
        if i not in str1:
            result = result + i
        
        elif str2.count(i) >1:
            result = result + i
            break
            
    return f'find_the_difference({str1, str2}) => {result}'

print(find_the_difference('abcd', 'abcde'))
print(find_the_difference('', 'y'))
print(find_the_difference('ae', 'aea'))
```

```python
'''Q.2 Given a function that accepts unlimited arguments, check and count how many data types are in those arguments. 
Finally return the total in a list List order is: [int, str, bool, list, tuple, dictionary]'''

def count_datatypes(*data):
    '''Count [int, str, bool, list, tuple, dictionary] in a unlimited arguments'''
    
    int_count = 0
    str_count = 0
    bool_count = 0
    list_count = 0
    tuple_count = 0
    dict_count = 0
    
    for i in data:
        if type(i) == int:
            int_count += 1
        elif type(i) == str:
            str_count += 1
        elif type(i) == bool:
            bool_count += 1
        elif type(i) == list:
            list_count += 1
        elif type(i) == tuple:
            tuple_count += 1
        elif type(i) == dict:
            dict_count += 1
            
    list1 = [int_count, str_count, bool_count, list_count, tuple_count, dict_count]
    return f'count_datatype({data}) => {list1}'
    
print(count_datatypes(1,45,'Hi',False))
print(count_datatypes([10,20],('t','ok'),2,3,1))
print(count_datatypes(4,21,('ES','EN'),('a','b'),False,[1,2,3],[4,5,6]))
```

```python
'''Q.3 A Fibonacci string is a precedence of the Fibonacci series. It works with any two characters of the English alphabet
(as opposed to the numbers 0 and 1 in the Fibonacci series) as the initial items and concatenates them together
as it progresses in a similar fashion as the Fibonacci series.'''

def fib_str(n, list_):
    
    a = list_[0]
    b = list_[1]
    
    list1 = []
    list1.append(a)
    list1.append(b)
    
    for i in range(n-2):
        a,b = b,b+a
        list1.append(b)
        
    res = ','.join(list1) 
    return f'fib_str({n}, {list_}) => {res}'
        
print(fib_str(3, ['j','h']))
print(fib_str(5, ['e','a']))
print(fib_str(6, ['n','k']))
```

```python
'''Q.4 Given an integer between 0 and 26, make a variable (self.answer). That variable would be assigned to a string
in this format: nines:your answer, threes:your answer, ones:your answer 
You need to find out how many ones, threes, and nines it would at least take for the number of each to add up to the 
given integer when multiplied by one,three or nine (depends).'''

def one_three_nine(number):
    '''Number to find out how many one, three and nine'''
    
    list1 = []
    if number >= 9:
        qnt1 = number//9
        rem1 = number%9
        if rem1 >= 3:
            qnt2 = rem1//3
            rem2 = rem1%3

        else:
            qnt2 = 0
    else:
        qnt1 = 0
        rem1 = number
        if rem1 >= 3:
            qnt2 = rem1//3
            rem2 = rem1%3
            
    rem3 = number - 9*qnt1 - 3*qnt2
    
    return f'one_three_nine({number}) => nines: {qnt1},threes: {qnt2},ones: {rem3}'

print(one_three_nine(10))
print(one_three_nine(15))
print(one_three_nine(22))
```

```python
'''Q.5 The Fibonacci sequence is a classic use case for recursive functions since the value of the sequence at a given index 
is dependent on the sum of the last two values. However, the recursion tree created by solving the Fibonacci sequence 
recursively can grow quite fast. Therefore it can be important to think about the implications of running a function 
recursively. Depending on the size of n needed and the capabilities of the system in question you might want to take a 
different approach.Write a non-recursive function that takes an integer n and returns the Fibonacci sequences value at index n.'''

def fib(n):

    a = 0
    b = 1
    for i in range(n):
        a,b = b, a+b

    return f'fib({n}) => {a}'
    
print(fib(1))
print(fib(2))
print(fib(6))
```
